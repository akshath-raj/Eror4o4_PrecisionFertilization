```python
import cv2
from ultralytics import YOLO

# Load the YOLO model
model = YOLO('/Users/akshathr/Downloads/best (4).pt')

# Open the video file
video_path = '/Users/akshathr/Downloads/WhatsApp Image 2024-07-23 at 8.49.00 PM [MConverter.eu].mp4'
cap = cv2.VideoCapture(video_path)

# Get the video writer initialized to save the output video
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Use 'mp4v' codec for .mp4 files
out = cv2.VideoWriter('/Users/akshathr/Downloads/output.mp4', fourcc, 20.0, (int(cap.get(3)), int(cap.get(4))))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Perform object detection
    results = model(frame)
    detections = results[0].boxes  # Get detections from results

    # Get frame dimensions
    height, width, _ = frame.shape

    # Define the height and width for the rectangles
    rect_height = 50
    rect_width = (width - 40) // 3  # Leaving space between rectangles
    rect_y = height - rect_height - 100  # Higher placement above the bottom

    # Initialize colors for the rectangles
    colors = [(255, 255, 255)] * 3

    # Draw bounding boxes, labels, and center points
    for box in detections:
        x1, y1, x2, y2 = map(int, box.xyxy[0])  # Extract box coordinates
        confidence = box.conf[0]  # Extract confidence score
        label = box.cls[0]  # Extract class label

        # Draw the bounding box
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

        # Draw the label and confidence score
        text = f'{model.names[int(label)]} {confidence:.2f}'
        cv2.putText(frame, text, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

        # Calculate the center point
        center_x = (x1 + x2) // 2
        center_y = (y1 + y2) // 2

        # Draw the center point
        cv2.circle(frame, (center_x, center_y), 10, (0, 0, 255), -1)

        # Check for collision with each rectangle
        for i in range(3):
            rect_x = i * (rect_width + 20)
            if rect_x <= center_x <= rect_x + rect_width and rect_y <= center_y <= rect_y + rect_height:
                colors[i] = (0, 0, 255)  # Change color to red if center point coincides

    # Draw the three rectangles at the bottom of the frame
    for i in range(3):
        rect_x = i * (rect_width + 20)  # Leave a space of 20 pixels between rectangles
        cv2.rectangle(frame, (rect_x, rect_y), (rect_x + rect_width, rect_y + rect_height), colors[i], -1)
        cv2.rectangle(frame, (rect_x, rect_y), (rect_x + rect_width, rect_y + rect_height), (0, 0, 0), 2)

    # Put the labels inside the rectangles
    cv2.putText(frame, 'n1', (rect_width // 2 - 10, rect_y + rect_height // 2 + 10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2)
    cv2.putText(frame, 'n2', (rect_width + rect_width // 2 - 10 + 20, rect_y + rect_height // 2 + 10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2)
    cv2.putText(frame, 'n3', (2 * rect_width + rect_width // 2 - 10 + 40, rect_y + rect_height // 2 + 10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2)

    # Write the frame to the output video file
    out.write(frame)

    # Display the frame (optional, you can remove this part if you don't need to display the video)
    cv2.imshow('Frame', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and writer objects
cap.release()
out.release()
cv2.destroyAllWindows()
```
